# Exploratory Data Analysis (EDA)

This notebook explores the Spotify Global Music Dataset (2009–2025) to understand relationships between audio features and track popularity.

Our goal is to identify patterns that may help predict `track_popularity`.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python
df = pd.read_csv("../data/spotify_data_clean.csv")

df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['track_popularity'].describe()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(numeric_only=True), cmap="coolwarm", center=0)
plt.title("Correlation Matrix")
plt.show()
```

```python
# Hist of Track Popularity
spot_dat = df

plt.figure()
plt.hist(pd.to_numeric(spot_dat['track_popularity'], errors='coerce'), bins=25)
plt.xlabel("Track Popularity")
plt.ylabel("Frequency")
plt.title("Track Popularity Distribution")
plt.show()
```

```python
# Track popularity vs Artist popularity

plt.figure()
plt.scatter(
    pd.to_numeric(spot_dat['artist_popularity'], errors='coerce'),
    pd.to_numeric(spot_dat['track_popularity'], errors='coerce'),
    alpha=0.5
)
plt.xlabel("Artist Popularity")
plt.ylabel("Track Popularity")
plt.title("Artist vs Track Popularity")
plt.show()
```

```python
#Followers vs Track popularity

plt.figure()
plt.scatter(
    pd.to_numeric(spot_dat['artist_followers'], errors='coerce'),
    pd.to_numeric(spot_dat['track_popularity'], errors='coerce'),
    alpha=0.5
)
plt.xlabel("Artist Followers")
plt.ylabel("Track Popularity")
plt.title("Followers vs Track Popularity")
plt.show()
```

```python
#Track lenght vs popularity

plt.figure()
plt.scatter(
    pd.to_numeric(spot_dat['track_duration_min'], errors='coerce'),
    pd.to_numeric(spot_dat['track_popularity'], errors='coerce'),
    alpha=0.5
)
plt.xlabel("Track Duration (min)")
plt.ylabel("Track Popularity")
plt.title("Length vs Popularity")
plt.show()
```

```python
# Popularity, explicit vs not explicit

plt.figure()
sns.boxplot(
    x=spot_dat['explicit'].astype(bool),
    y=pd.to_numeric(spot_dat['track_popularity'], errors='coerce')
)
plt.xlabel("Explicit")
plt.ylabel("Track Popularity")
plt.title("Popularity, Not explicit vs Explicit")
plt.show()
```

```python
#Popularity by album type

plt.figure()
sns.boxplot(
    x=spot_dat['album_type'],
    y=pd.to_numeric(spot_dat['track_popularity'], errors='coerce')
)
plt.xlabel("Album Type")
plt.ylabel("Track Popularity")
plt.title("Popularity by Album type")
plt.show()
```

```python
# Album size vs Popularity

plt.figure()
plt.scatter(
    pd.to_numeric(spot_dat['album_total_tracks'], errors='coerce'),
    pd.to_numeric(spot_dat['track_popularity'], errors='coerce'),
    alpha=0.5
)
plt.xlabel("Total Tracks in Album")
plt.ylabel("Track Popularity")
plt.title("Album Size vs Popularity")
plt.show()
```

```python
# Avg track popularity by year of release

spot_dat['album_release_date'] = pd.to_datetime(
    spot_dat['album_release_date'], errors='coerce'
)

spot_dat['release_year'] = spot_dat['album_release_date'].dt.year

# Group and summarize
year_summary = (
    spot_dat
    .groupby('release_year')['track_popularity']
    .mean()
    .reset_index(name='avg_pop')
)

# Plot
plt.figure()
plt.plot(year_summary['release_year'], year_summary['avg_pop'])
plt.xlabel("Release Year")
plt.ylabel("Average Popularity")
plt.title("Average popularity of tracks released each year")
plt.show()
```
